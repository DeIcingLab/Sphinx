# SPHINx project

SPHINx is a french DGAC research and development project:

" **S**olutions **P**erformantes à **H**aut niveau d’**I**ntégration et coûts/masses réduits de **N**acelle et mât UHBR »


## New composite technologies

The SPHINX project aims to create composite technologies to lighten the structural components needed to develop UHBR (Ultra High Bypass Ratio) engines. These new engines will deliver fuel savings of up to 10%. 
![Hutchinson](https://www.hutchinson.com/sites/default/files/news/2024-10/photo_mat_reacteur.png)

## De-icing composite structures

Jet engine air intakes are typically made of metal in order to withstand the high temperatures of hot air anti-icing systems. Other ice protection solutions need to be explored if the use of composites is to be extended to these critical areas.
