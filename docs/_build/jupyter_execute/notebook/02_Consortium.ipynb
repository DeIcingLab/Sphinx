# Members of the electromechanical deicing WP

## Airbus

The SPHINx project is led by [Airbus](https://aircraft.airbus.com/fr), which is responsible for:
- developing specifications (general objectives, interfaces, integration data, performance/weight/cost targets, industrial constraints, etc.)
- monitoring and coordinating partners
- selecting options and solutions
- validating solutions against requirements, including functional validation through testing.

## Institut Clément Ader 

The [Institut Clément Ader](https://ica.cnrs.fr/) is responsible for:
- the development and feasibility of low-consumption electromechanical de-icing systems
- analysis and optimization for the integration of systems on composite substrates
- the definition of a prototype for testing and measurements in an icing wind tunnel

![Activités ICA](./figures/ICAwork.png)  

## ISAE Supaero and CRITT Mécanique et Composites

ISAE Supaero (https://www.isae-supaero.fr/) and CRITT Mécanique et Composites (https://www.mecanique-composite.com/) are supporting the Clément Ader Institute in this project by helping to build the prototypes.
